```python
import pandas as pd
netflix_data = pd.read_csv('../data/netflix_titles.csv')
```

```python
netflix_data.head()
```

```python
## Visualize the number of movies grouped by each genre.
genre_counts = netflix_data['listed_in'].str.split(',').explode().value_counts()
print(genre_counts)
```

```python
genre_counts.head(10).plot(kind='bar', title='Top 10 most popular genres')
```

```python
release_counts = netflix_data['release_year'].value_counts().sort_index()
print(release_counts)
```

```python
release_counts.plot(kind='line', title='Releases by Year')
```

```python
rating_counts = netflix_data['rating'].value_counts()
print(rating_counts)
```

```python
rating_counts.plot(kind='bar', title='Rating distribution')
```

```python
## Movies per country
country_counts = netflix_data['country'].str.split(',').explode().value_counts()
print(country_counts)
```

```python
country_counts.head(10).plot(kind='bar', title='Top 10 countrys with most movies')
```
